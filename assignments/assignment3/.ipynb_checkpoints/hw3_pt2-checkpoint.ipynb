## Problem 1 (20pt)

For the digital communication system in part 1, problem 1.

1. Implement both MAP and MLE rule as two functions. (10pt)
2. Plot the probability of error given  $P(A_0)$ range from [0,1] under MAP and MLE rules, using different colors. (10pt)

```python
# import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
from tqdm import tqdm

def Mle(observation, P, PA0):
    return np.argmax(P[:, observation])

def Map(observation,P,PA0):
    scaled_apps = P[:,observation] * np.array([PA0, 1-PA0])
    #denominator is the same for both
    return np.argmax(scaled_apps)
 
P = np.array([[1/2,1/4,1/4],[1/10,3/10,6/10]])

#Perror for MLE is total probability of the sum of taking "wrong" decisions
def sim2to3(rule,P,PA0):
    PA1 = 1-PA0
    trials = int(1e5)
    count_error = 0
    for _ in range(trials):
        #pick an input according to some probability
        x = npr.choice(a = [0,1],
                      size = 1,
                      replace = True,
                      p = [PA0,PA1])
        x = int(x)
        #send an output based on x
        y = npr.choice(a = [0,1,2],
                      size = 1,
                      replace = True,
                       p = P[int(x),:])
        y = int(y)
        #make a decision on the observed 
        guess = rule(y,P,PA0)
        guess = int(guess)
        #error
        if guess != x:
            count_error += 1
    
    return count_error/trials


inputSet = np.linspace(0,1,21)

pe_ML = [sim2to3(Mle,P,PA0) for PA0 in inputSet]
pe_MAP = [sim2to3(Map,P,PA0) for PA0 in inputSet]

plt.plot(inputSet, pe_ML,color = 'red', linestyle = '-',label = "ML")
plt.plot(inputSet,pe_MAP, linestyle = ':', color = 'green',label = "MAP")
plt.xlabel("P(A_0)")
plt.ylabel("P(Error)")
plt.title("P(Error) v. Decision")
plt.legend()
plt.show()
```

## problem 2 (10pt)

A pharmaceutical company is testing a new vaccine against a virus. Based on previous studies, the vaccine is known to be 70% effective, meaning that each vaccinated person has a 70% chance of developing immunity.

In a trial, 50 people receive the vaccine.

Simulate the experiment by generating 1,000 random trials and estimate the probability of at least 35 people developing immunity.

```python
trials = 1_000
n = 50
desireSuccess = 35
results = npr.binomial(n,p = 0.7, size = trials)
print(f"P of at least 35 people achieving immunity = {np.mean(results>=desireSuccess)}")

    
```

## Problem 3 (20pt)

A small coffee shop tracks the number of customers arriving per hour. Based on past data, they believe the arrival rate  $\lambda$
follows one of two possible values:
* Hypothesis 0 ( $H_0$ ): The arrival rate is 5 customers per hour ( $\lambda_0 = 5$ ).
* Hypothesis 1 ( $H_1$): The arrival rate is 10 customers per hour ( $\lambda_1 = 10$ ).

Before observing any data, the shop owner believes that both hypotheses are equally likely:

$P(H_1) = P(H_0) = 0.5$


Given an observed customer count  X = k  in one hour, use MAP estimation to determine which hypothesis is more probable. 

- Implement a python code to decide, given any k, output the decided hypothesis

```python
import scipy.stats as stats
#P(H_1|kcustomers)
def experiment(k):
    #P(X = K|H_i)
    pIfH0 = stats.poisson.pmf(k,mu = 5)
    pIfH1 = stats.poisson.pmf(k,mu = 10)
    #P(H_i)
    pH0 = 0.5
    pH1 = 0.5
    #P(K) will be the same for both
    #P(H_i|X = K)
    pH0_appscale = 0.5 * pIfH0
    pH1_appscale = 0.5 * pIfH1
    implies = np.argmax([ pH0_appscale , pH1_appscale  ])
    pImplies = np.max([ pH0_appscale , pH1_appscale  ])
    option = [5,10]
    print()
    print(f"The more probable hypothesis is H{implies}, that {option[implies]} customers per hour")
    print(f"from observing X = {k} customers in an hour")
    print(f"With a probability of p = {pImplies: .3f}")
    print()
    return

for x in range(21):
    experiment(x)

    
```
